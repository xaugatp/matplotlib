```python
 # A scatter plot is a type of plot or mathematical diagram using catesian coordinates
 # to display values for typically two variables for a set of data.

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.scatter(x,y)
plt.show()
```

```python
''' 
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

'''
```

```python
''' 
The 'subplot()' function is in matplotlib is used to create multiple plots within 
a single figure. It allows you to arrange plots in a grid-like structure, specifying the number of rows and columns. This is particularly useful when you want to compare different datasets or visualize multiple aspects of the same data.

'''
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, plot 1
plt.plot(x1, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)  # 2 rows, 2 columns, plot 2
plt.plot(x2, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)  # 2 rows, 2 columns, plot 3
plt.plot(x3, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)  # 2 rows, 2 columns, plot 4
plt.plot(x4, y4)
plt.title('Line Plot 4')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```

```python
''' 

Box Plot:
A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the summary statistics of a dataset, such as the minimum, first quartile (Q1), median, third quartile (Q3), and maximum. The box plot provides a visual summary of the central tendency, spread, and skewness of the data, making it useful for identifying outliers and understanding the overall distribution.

Why is it used?
Box plots are used for the following purposes:

Visualizing Distribution: Box plots provide a concise way to visualize the distribution of data, highlighting key statistical measures.
Comparing Distributions: They are effective for comparing the central tendency and spread of multiple datasets.
Identifying Outliers: Outliers, which are data points significantly different from the rest of the dataset, can be easily identified in a box plot.
'''
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data for box plots
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating a box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot Example')
plt.show()
```
